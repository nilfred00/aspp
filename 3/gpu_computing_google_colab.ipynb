```python
import numpy as np
import cupy as cp
import timeit
```

```python
# Loop over matrix sizes
for N in [128,256,512,1024,2048]:

  # Create matrices
  matrix_np = np.random.rand(N,N)
  matrix_cp = cp.asarray(matrix_np)


  string = f"========== Matrix: {N} x {N} =========="
  print(string)

  # Fourier transforms
  print("NumPy:")
  %timeit np.fft.fft2(matrix_np)

  print("CuPy:")
  %timeit cp.fft.fft2(matrix_cp)

  print("="*len(string))


```

```python
# Loop over matrix sizes
for N in [128,256,512,1024,2048]:

  # Create matrices
  matrix_np = np.random.rand(N,N)
  matrix_np = np.array(matrix_np, dtype=np.float32)
  matrix_cp = cp.asarray(matrix_np)


  string = f"========== Matrix: {N} x {N} =========="
  print(string)

  # Fourier transforms
  print("NumPy:")
  %timeit np.fft.fft2(matrix_np)

  print("CuPy:")
  %timeit cp.fft.fft2(matrix_cp)

  print("="*len(string))


```
